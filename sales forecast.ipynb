```python
def parse(data_dir, filename):
    name = ''
    data = []
    datafile = os.path.join(data_dir, filename)
    with open(datafile, 'rb') as f:
        datareader = csv.reader(f)
        
        header_row = datareader.next()
        print(header_row)
        
        for row in datareader:
            data.append(row)
    
    pprint.pprint(data[:100])        
    return data          
```
